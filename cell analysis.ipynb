```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
clump_sizes = pd.read_csv('data.csv').values
clump_sizes
```

## Interpreting Results

### The initial distribution is skewed towards most clumps being in the 0-10 range. 
These were probably simply pixels that the algorithm took a liking to, too small for any noticeable change in the image.

```python
sns.displot(clump_sizes, legend=False)
plt.show()
```

### Here is the weighted distribution, giving higher counts for higher values. This resolves the skew issue.

```python
weighted_clump_sizes = sum([list(c)*int(c) for c in clump_sizes], [])
median, mean = np.median(weighted_clump_sizes), np.mean(weighted_clump_sizes)
sns.displot(weighted_clump_sizes, bins=30)
plt.axvline(median, c='red');
plt.show()
```

```python
grid_pixels = [966264, 930240, 936054, 886422] # the grid pixels for each of the sampled grids from concentration 1
one_cell = mean/np.mean(grid_pixels)  # on average, a single food source cell makes up this ratio of a 1x1 mm square 
```
