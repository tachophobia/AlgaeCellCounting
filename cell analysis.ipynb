```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
clump_sizes = np.concatenate([pd.read_csv('1tr.csv').values, pd.read_csv('1br.csv').values, pd.read_csv('1bl.csv').values])
clump_sizes
```

## Interpreting Results

The distribution cuts off at 10 since the minimum area required for an algae cell to be considered is an arbitrary 10 pixels

```python
import numpy as np
sns.displot(clump_sizes, legend=False, bins=30)
plt.axvline(np.median(clump_sizes), c='red')
plt.show()
```

```python
grid_pixels = [966264] # the grid pixels for each of the sampled grids
one_cell = np.mean(clump_sizes)/np.mean(grid_pixels)  # on average, a single food source cell makes up this ratio of a 1x1 mm square 
```
